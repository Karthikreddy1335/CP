```python
# Subset Sum using Backtracking

def subset_sum(arr, n, target):

    # If target becomes 0, subset is found
    if target == 0:
        return True

    # If no elements left or target becomes negative
    if n == 0 or target < 0:
        return False

    # Choice 1: include current element
    include = subset_sum(arr, n - 1, target - arr[n - 1])

    # Choice 2: exclude current element
    exclude = subset_sum(arr, n - 1, target)

    return include or exclude


# Read input
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

# Check subset sum
result = subset_sum(arr, n, target)

# Print result
print(result)
```
